# Algoritmos Genéticos

## Introducción

Los algoritmos genéticos (AG) son una clase de algoritmos de optimización y búsqueda basados en los principios de la genética y la selección natural de Darwin. Forman parte de los algoritmos evolutivos y se utilizan para resolver problemas complejos donde los métodos tradicionales pueden ser ineficientes o inadecuados. Su capacidad para explorar grandes espacios de búsqueda y encontrar soluciones óptimas o cercanas al óptimo los hace valiosos en diversas aplicaciones.

## ¿Qué es un algoritmo genético?

Un algoritmo genético es un método de optimización inspirado en el proceso de evolución biológica. Emula la forma en que las poblaciones de organismos evolucionan a través de generaciones mediante mecanismos de selección natural, cruce y mutación. El objetivo es encontrar la mejor o las mejores soluciones a un problema específico, representado mediante una función de aptitud que evalúa la calidad de cada solución candidata.

## ¿Cómo funciona un algoritmo genético?

El funcionamiento de un algoritmo genético se basa en los siguientes pasos fundamentales:

1. **Inicialización de la población**: Se crea una población inicial de soluciones candidatas (individuos) de forma aleatoria o utilizando conocimientos previos.

2. **Evaluación**: Cada individuo de la población es evaluado mediante la función de aptitud, que asigna un valor numérico representando su calidad o adaptación al problema.

3. **Selección**: Se eligen los individuos más aptos para reproducirse. Los métodos de selección pueden ser proporcionales a la aptitud, torneos, entre otros.

4. **Cruce (recombinación)**: Los individuos seleccionados (padres) se combinan para producir nuevos individuos (descendientes), intercambiando partes de su información genética. Esto permite explorar nuevas áreas del espacio de búsqueda.

5. **Mutación**: Se aplican cambios aleatorios a los descendientes para mantener la diversidad genética y evitar la convergencia prematura hacia soluciones subóptimas.

6. **Reemplazo**: Los nuevos individuos sustituyen parcial o totalmente a la población actual, formando una nueva generación.

7. **Iteración**: Se repiten los pasos de evaluación, selección, cruce y mutación durante varias generaciones hasta cumplir con un criterio de parada, como alcanzar un número máximo de generaciones o una aptitud satisfactoria.

## Componentes clave de un algoritmo genético

- **Población**: Conjunto de individuos que representan posibles soluciones al problema. El tamaño de la población afecta la diversidad y la convergencia del algoritmo.

- **Representación genética**: Forma en que se codifican las soluciones, comúnmente como cadenas de bits, números reales, permutaciones, entre otros, dependiendo del problema.

- **Función de aptitud**: Mide la calidad de cada individuo en relación con el objetivo del problema. Es esencial para guiar el proceso evolutivo hacia mejores soluciones.

- **Operadores genéticos**:
  - **Selección**: Mecanismo para elegir los individuos que participarán en la reproducción. Favorece a los más aptos, pero permite que individuos menos aptos también tengan oportunidad para mantener la diversidad.
  - **Cruce**: Combina información genética de dos o más padres para producir descendientes. Métodos comunes incluyen cruce de un punto, de dos puntos y uniforme.
  - **Mutación**: Introduce variaciones aleatorias en los individuos. Ayuda a explorar nuevas soluciones y a evitar la estancamiento en óptimos locales.

- **Parámetros del algoritmo**: Incluyen la tasa de cruce, tasa de mutación, tamaño de la población y criterios de parada, que deben ser ajustados cuidadosamente para el problema en cuestión.

## ¿Cómo aprende un algoritmo genético?

El algoritmo genético aprende y mejora las soluciones a través de un proceso iterativo que simula la evolución natural:

- **Selección de los más aptos**: Al favorecer la reproducción de individuos con mayor aptitud, el algoritmo incrementa la probabilidad de que las características beneficiosas se transmitan a las siguientes generaciones.

- **Recombinación de información**: El cruce permite combinar las mejores características de diferentes individuos, explorando nuevas soluciones que podrían ser superiores a las anteriores.

- **Introducción de variabilidad**: La mutación añade diversidad genética, permitiendo al algoritmo escapar de óptimos locales y explorar regiones inexploradas del espacio de búsqueda.

- **Adaptación progresiva**: A lo largo de las generaciones, la población se adapta al problema, mejorando gradualmente la calidad de las soluciones hasta alcanzar o aproximarse al óptimo global.

## Aplicaciones de los algoritmos genéticos

Los algoritmos genéticos se han aplicado exitosamente en diversos campos:

- **Optimización de funciones complejas**: Donde las funciones no son derivables o presentan múltiples óptimos locales.

- **Diseño y ingeniería**: Optimización de estructuras, circuitos electrónicos, diseño aerodinámico, entre otros.

- **Inteligencia artificial y aprendizaje automático**: Entrenamiento de redes neuronales, selección de características y ajuste de hiperparámetros.

- **Problemas combinatorios**: Como el problema del viajante, planificación de rutas y programación de tareas.

- **Bioinformática**: Análisis de secuencias genéticas, predicción de estructuras proteicas y estudios evolutivos.

## Ventajas de los algoritmos genéticos

- **Robustez y flexibilidad**: Pueden aplicarse a una amplia gama de problemas sin requerir información específica del gradiente o derivadas.

- **Búsqueda global**: Su capacidad para explorar múltiples regiones del espacio de búsqueda reduce la probabilidad de quedar atrapados en óptimos locales.

- **Adaptabilidad**: Pueden ajustarse y combinarse con otros métodos para mejorar su rendimiento en problemas específicos.

## Limitaciones de los algoritmos genéticos

- **Alto costo computacional**: Requieren evaluar muchas soluciones en cada generación, lo que puede ser costoso para funciones de aptitud complejas.

- **Parámetros sensibles**: El rendimiento depende de la configuración adecuada de los parámetros, lo cual puede requerir experimentación.

- **No garantizan optimalidad**: Aunque pueden aproximarse al óptimo global, no siempre lo aseguran, especialmente en problemas muy complejos.

# 1: Diseño de un Algoritmo Genético para Camuflaje de un Cubo en un Entorno

## Introducción

El objetivo es desarrollar un algoritmo genético que ajuste el color de un cubo para que se camufle con el entorno, específicamente, que su color en escala de grises se acerque al del fondo. Para lograr esto, utilizaremos una población inicial y aplicaremos los principios de los algoritmos genéticos para evolucionar hacia la solución óptima.

La población tendrá un tamaño del 100% (se asume que es un número fijo, por ejemplo, 100 individuos), y en cada generación, solo sobrevivirán el 20% de los individuos cuya representación de color esté más cercana al valor de gris del fondo. Este proceso se repetirá hasta que el cubo alcance un nivel de camuflaje satisfactorio.

## Componentes del Algoritmo Genético

Para resolver este problema, es necesario definir claramente cada uno de los componentes del algoritmo genético:

### 1. Representación Genética (Codificación)

- **Genotipo**: Cada individuo en la población representa un posible color del cubo.
- **Fenotipo**: El color visible del cubo en escala de grises.
- **Codificación**: Utilizaremos un valor numérico que representa el nivel de gris, donde 0 es negro y 255 es blanco.

### 2. Población Inicial

- **Tamaño de la Población**: 100 individuos.
- **Inicialización**: Los valores de gris de los individuos se generan aleatoriamente entre 0 y 255.

### 3. Función de Aptitud (Fitness Function)

- **Objetivo**: Minimizar la diferencia absoluta entre el nivel de gris del individuo y el nivel de gris del fondo.
- **Función de Aptitud**: \( \text{Aptitud}(i) = -|\text{Gris}_{\text{individuo}} - \text{Gris}_{\text{fondo}}| \)
- **Nota**: Al usar el valor negativo de la diferencia, los individuos con menor diferencia tendrán mayor aptitud.

### 4. Selección

- **Criterio de Supervivencia**: Solo el 20% de los individuos con mayor aptitud sobreviven para la siguiente generación.
- **Método de Selección**: Selección elitista, donde se conservan los mejores individuos.

### 5. Operadores Genéticos

#### a. Cruce (Crossover)

- **Tipo de Cruce**: Cruce uniforme.
- **Proceso**: Se seleccionan pares de padres y se combinan sus valores de gris para producir descendientes.

#### b. Mutación

- **Probabilidad de Mutación**: Pequeña, por ejemplo, 1%.
- **Proceso**: Se añade o resta un valor aleatorio pequeño al nivel de gris del individuo.

### 6. Criterio de Parada

- **Condiciones**:
  - Alcanzar un nivel de aptitud satisfactorio (por ejemplo, una diferencia de gris menor a un umbral).
  - Llegar a un número máximo de generaciones.

## Pasos del Algoritmo Genético

1. **Inicialización**: Crear una población inicial de 100 individuos con valores de gris aleatorios entre 0 y 255.

2. **Evaluación de la Aptitud**: Calcular la aptitud de cada individuo utilizando la función de aptitud definida.

3. **Selección de Sobrevivientes**: Retener el 20% de los individuos con mayor aptitud.

4. **Generación de Nueva Población**:
   - **Cruce**: Utilizar los individuos sobrevivientes para generar nuevos individuos mediante cruce.
   - **Mutación**: Aplicar mutación a algunos de los nuevos individuos para mantener la diversidad genética.

5. **Reemplazo**: La nueva población reemplaza a la antigua, manteniendo el tamaño total en 100 individuos.

6. **Verificación del Criterio de Parada**: Si se cumple el criterio de parada, finalizar el algoritmo; de lo contrario, volver al paso 2.

## 3. Implementación

````python
import random
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import seaborn as sns

# Parámetros del algoritmo genético
TAMANO_POBLACION_INICIAL = 100  # Número total de individuos en la población inicial.
TAMANO_POBLACION_APTA_OBJETIVO = 95  # Define el tamaño de la población apta para el objetivo.
PORCENTAJE_SOBREVIVIENTES = 0.2  # Porcentaje de individuos que sobreviven en cada generación.
PROB_MUTACION = 0.01  # Probabilidad de que un individuo sufra una mutación.
UMBRAL_APTITUD = 2  # Diferencia mínima aceptable entre un individuo y el fondo para detener el algoritmo.
MAX_GENERACIONES = 1000  # Número máximo de generaciones permitidas en el algoritmo.
GRIS_FONDO = 150  # Valor del fondo en escala de grises, usado como objetivo estático.


def crear_poblacion_inicial(tamano):
    """
    Crea una población inicial de individuos representados por valores enteros
    entre 0 y 255, que corresponden a colores en escala de grises.

    Parámetros:
        - tamano (int): Número de individuos en la población inicial.

    Retorna:
        - poblacion (list): Lista de valores enteros que representan los colores iniciales.
    """
    poblacion = []
    for _ in range(tamano):
        individuo = random.randint(0, 255)
        poblacion.append(individuo)
    return poblacion

def calcular_aptitud(individuo, gris_fondo):
    """
    Calcula la aptitud de un individuo en función de su proximidad al color del fondo.

    Parámetros:
        - individuo (int): Valor en escala de grises del individuo (0 a 255).
        - gris_fondo (int): Valor en escala de grises del fondo (0 a 255).

    Retorna:
        - aptitud (int): Valor negativo de la diferencia absoluta entre el individuo y el fondo.
          Una menor diferencia da como resultado una mayor aptitud.

    Fórmula:
        aptitud = -|individuo - gris_fondo|

    """
    return -abs(individuo - gris_fondo)


def seleccionar_sobrevivientes(poblacion, aptitudes, porcentaje):
    """
    Selecciona un porcentaje de los individuos con las mejores aptitudes para 
    sobrevivir a la siguiente generación.

    Parámetros:
        - poblacion (list): Lista de individuos actuales.
        - aptitudes (list): Lista de valores de aptitud correspondientes a los individuos.
        - porcentaje (float): Porcentaje de la población que sobrevivirá (0 a 1).

    Retorna:
        - sobrevivientes (list): Lista de individuos seleccionados.

    """
    num_sobrevivientes = int(len(poblacion) * porcentaje)  # Se calcula el número de individuos que sobrevivirán

    individuos_aptitud = []  # Se inicializa una lista para almacenar individuos con sus aptitudes
    for i in range(len(poblacion)): 
        # Se combina cada individuo con su aptitud en una tupla
        individuos_aptitud.append((poblacion[i], aptitudes[i]))

    # Se ordena la lista de tuplas por aptitud en orden descendente
    individuos_aptitud.sort(key=lambda x: x[1], reverse=True)

    sobrevivientes = []  # Lista para almacenar los individuos seleccionados
    for i in range(num_sobrevivientes):  
        sobrevivientes.append((individuos_aptitud[i][0],individuos_aptitud[i][1])) # Se añaden los mejores individuos a la lista de sobrevivientes
    return sobrevivientes 


def cruzar(padre1, padre2):
    """
    Realiza un cruce entre dos padres para generar un nuevo individuo.
    En este caso, se utiliza un promedio ponderado entre los valores de los padres.

    Parámetros:
        - padre1 (int): Valor en escala de grises del primer padre.
        - padre2 (int): Valor en escala de grises del segundo padre.

    Retorna:
        - hijo (int): Nuevo valor en escala de grises calculado como un promedio ponderado.
    """
    # Genera un peso aleatorio entre 0 y 1 para ponderar los padres
    peso = random.random()
    # Calcula el valor del hijo como un promedio ponderado entre los dos padres
    hijo = int(peso * padre1 + (1 - peso) * padre2)
    # Devuelve el valor del hijo
    return hijo

def mutar(individuo):
    """
    Introduce una pequeña variación aleatoria en el valor del individuo con
    cierta probabilidad.

    Parámetros:
        - individuo (int): Valor en escala de grises del individuo (0 a 255).

    Retorna:
        - individuo_mutado (int): Nuevo valor del individuo, con posible mutación.

    """
    if random.random() < PROB_MUTACION:  # Se genera un número aleatorio y verifica si está dentro del umbral de mutación
        individuo += random.randint(-100, 100)  # Se ajusta el valor del individuo aleatoriamente en un rango pequeño
        individuo = max(0, min(255, individuo))  # Se asegura que el valor permanezca dentro del rango válido (0 a 255)
    return individuo  # Se devuelve el individuo (mutado o no)

def generar_nueva_poblacion(sobrevivientes, TAMANO_POBLACION_INICIAL):
    """
    Genera una nueva población a partir de los sobrevivientes mediante cruce y mutación.

    Parámetros:
        - sobrevivientes (list): Lista de individuos seleccionados para sobrevivir.
        - TAMANO_POBLACION_INICIAL (int): Tamaño total de la población a generar.

    Retorna:
        - nueva_poblacion (list): Lista con la nueva población generada.

    """
    nueva_poblacion = []  # Se inicializa una lista para almacenar la nueva población (solamente individuos)
    for individuo, aptitud in sobrevivientes:
        nueva_poblacion.append(individuo)
         
    while len(nueva_poblacion) < TAMANO_POBLACION_INICIAL:  # Se itera hasta completar el tamaño de la población inicial
        padres = random.sample(sobrevivientes, 2)  # Se selecciona dos padres aleatorios de los sobrevivientes
        hijo = cruzar(padres[0][0], padres[1][0])  # Se genera un nuevo individuo mediante cruce
        hijo = mutar(hijo)  # Se aplica mutación al nuevo individuo
        nueva_poblacion.append(hijo)  # Se añade el nuevo individuo a la población
    return nueva_poblacion 


# Algoritmo genético para el caso base
def algoritmo_genetico_base():
    """
    Implementa un algoritmo genético básico para optimizar el color de un cubo en escala de grises
    para que coincida con un fondo objetivo (GRIS_FONDO).

    El algoritmo utiliza selección elitista, cruce, y mutación para evolucionar una población de colores
    hacia la solución óptima en un número limitado de generaciones.

    Variables Globales Utilizadas:
        - TAMANO_POBLACION_INICIAL: Número de individuos en la población inicial.
        - MAX_GENERACIONES: Número máximo de generaciones permitidas.
        - GRIS_FONDO: Color objetivo del fondo en escala de grises (0 a 255).
        - UMBRAL_APTITUD: Diferencia mínima entre el mejor individuo y el fondo para detener el algoritmo.
        - PORCENTAJE_SOBREVIVIENTES: Porcentaje de la población que sobrevive en cada generación.

    Devuelve:
        - mejor_por_generacion: Lista con el mejor individuo (color) de cada generación.
    """
    # Inicialización de la población inicial y estructuras para el análisis.
    poblacion = crear_poblacion_inicial(TAMANO_POBLACION_INICIAL)  # Se crea la población inicial con valores aleatorios.
    historial_mejor_individuo = []  # Lista para almacenar el mejor individuo por generación.
    historial_color_promedio = []  # Lista para almacenar el color promedio por generación.
    generacion = 0  # Contador de generaciones.
    individuos_aptos=0
    
    while generacion < MAX_GENERACIONES and individuos_aptos != TAMANO_POBLACION_APTA_OBJETIVO:
        # Se calculan las aptitudes de los individuos.
        aptitudes = []
        individuos_aptos_gen=0
        for ind in poblacion:
            aptitudes.append(calcular_aptitud(ind, GRIS_FONDO))  # Se evalua aptitud de cada individuo.
            diferencia_ind = abs(ind - GRIS_FONDO)
            if diferencia_ind <= UMBRAL_APTITUD:
                individuos_aptos_gen+=1
        if individuos_aptos_gen == TAMANO_POBLACION_APTA_OBJETIVO:
            individuos_aptos = individuos_aptos_gen
        prom_color = sum(poblacion) / len(poblacion)  # Se calcula el color promedio de la población.
        historial_color_promedio.append(prom_color)  # Se almacena el color promedio de la generación.
        
        ## Se encuentra el mejor individuo y estadísticas de aptitud.
        max_aptitud = max(aptitudes)  # Aptitud máxima en la generación.
        mejor_individuo = poblacion[aptitudes.index(max_aptitud)]  # Mejor individuo de la generación.

        # Se seleccionan los sobrevivientes y se genera una nueva población.
        sobrevivientes = seleccionar_sobrevivientes(poblacion, aptitudes, PORCENTAJE_SOBREVIVIENTES)
        poblacion = generar_nueva_poblacion(sobrevivientes, TAMANO_POBLACION_INICIAL)

        ## Se almacenan estadísticas de la generación actual.
        historial_mejor_individuo.append(mejor_individuo)
        generacion += 1
        
    if individuos_aptos == TAMANO_POBLACION_APTA_OBJETIVO:
        print(f"Se encontraron {TAMANO_POBLACION_APTA_OBJETIVO} individuos aptos en la generación {generacion}")
    else:
        print(f"Se alcanzó el límite de generaciones sin encontrar {TAMANO_POBLACION_APTA_OBJETIVO} individuos aptos")
    
    # Se grafican los resultados obtenidos.

    # Grafica resultados del nivel de gris del mejor individuo.
    plt.figure(figsize=(12, 6))
    plt.plot(historial_mejor_individuo, label='Mejor Individuo')  # Grafica mejor individuo por generación.
    plt.axhline(GRIS_FONDO, color='r', linestyle='--', label='Gris de Fondo')  # Línea representando el fondo objetivo.
    plt.title('Nivel de Gris por Generación')
    plt.xlabel('Generación')
    plt.ylabel('Nivel de Gris')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Grafica resultados del color promedio de la población.
    plt.figure(figsize=(12, 6))
    plt.plot(historial_color_promedio, label='Color Promedio')  # Grafica color promedio por generación.        
    plt.axhline(GRIS_FONDO, color='r', linestyle='--', label='Gris de Fondo')  # Línea representando el fondo objetivo.
    plt.title('Color Promedio por Generación')
    plt.xlabel('Generación')
    plt.ylabel('Color Promedio')
    plt.legend()
    plt.grid(True)
    plt.show()
    return historial_color_promedio
    
# Función para graficar el heatmap
def graficar_heatmap(historial_color_promedio, gris_fondo):
    data = np.array(historial_color_promedio).reshape(-1, 1)
    plt.figure(figsize=(8, 12))
    sns.heatmap(data, annot=False, cmap="Greys", cbar=True, xticklabels=False, 
                yticklabels=[f"Gen {i+1}" for i in range(len(historial_color_promedio))])
    plt.axhline(y=gris_fondo, color='red', linestyle='--', label='Gris Fondo')
    plt.title("Progreso del Color Promedio por Generación")
    plt.xlabel("Población")
    plt.ylabel("Generación")
    plt.legend()
    plt.show()

# Ejecutar el algoritmo y visualizar el progreso por generación
historial_color_promedio = algoritmo_genetico_base()
graficar_heatmap(historial_color_promedio, GRIS_FONDO)

````

````python
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib.colors import ListedColormap

def create_maze(dim):
    # Create a grid filled with walls (1)
    # Set the seed to ensure reproducibility
    np.random.seed(2)
    maze = np.ones((dim * 2 + 1, dim * 2 + 1), dtype=int)

    # Define the starting point
    x, y = (0, 0)
    maze[2 * x + 1, 2 * y + 1] = 0  # Mark start as open path

    # Initialize the stack with the starting point for DFS
    stack = [(x, y)]
    
    while stack:
        x, y = stack[-1]

        # Define possible directions (right, down, left, up)
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        random.shuffle(directions)  # Randomize order for more organic mazes

        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            # Check if next cell is within bounds and is a wall
            if 0 <= nx < dim and 0 <= ny < dim and maze[2 * nx + 1, 2 * ny + 1] == 1:
                # Break wall to create path
                maze[2 * nx + 1, 2 * ny + 1] = 0
                maze[2 * x + 1 + dx, 2 * y + 1 + dy] = 0
                stack.append((nx, ny))  # Move to the next cell
                break
        else:
            stack.pop()  # Backtrack if no unvisited neighbors

    # Create entrance and exit points
    maze[1, 0] = 0  # Entrance
    maze[-2, -1] = 0  # Exit

    return maze

def display_maze(maze):
    cmap = ListedColormap(['white', 'black', 'green', 'black'])
    plt.figure(figsize=(6, 6))
    plt.pcolor(maze[::-1], cmap=cmap, edgecolors='k', linewidths=2)
    plt.gca().set_aspect('equal')
    plt.xticks([])
    plt.yticks([])
    plt.title('Maze with Entrance and Exit')
    plt.show()

# Define the dimension of the maze (adjustable)
dim =7 # This can be any positive integer
maze = create_maze(dim)
display_maze(maze)
````

````python
def plot_solution_path_fixed(maze, start, end, visited_positions, generation, best_fitness):
    """
    Dibuja un mapa del laberinto resaltando el camino del mejor individuo.
    """
    maze_array = np.array(maze)
    plt.figure(figsize=(10, 8))
    plt.title(f"Solución encontrada en la generación {generation}\nFitness: {best_fitness}")
    
    # Dibujar el laberinto
    for y in range(len(maze_array)):
        for x in range(len(maze_array[0])):
            if maze_array[y][x] == 1:  # Pared
                plt.fill_between([x, x + 1], [y, y], [y + 1, y + 1], color="black")
            else:  # Celda transitable
                plt.fill_between([x, x + 1], [y, y], [y + 1, y + 1], color="white")
    
    # Dibujar la entrada 
    startx, starty = start
    plt.fill_between([startx, startx + 1], [starty, starty], [starty + 1, starty + 1], color="green")
        
    # Dibujar el camino visitado
    for pos in visited_positions:
        y, x = pos  # Posición actual
        plt.fill_between([x, x + 1], [y, y], [y + 1, y + 1], color="green")
    
    plt.gca().invert_yaxis()  # Invertir eje Y para que coincida con el laberinto
    plt.grid(visible=False)
    plt.show()
````

````python
import random

def generate_genome(genome_length):
    """
    Generates a genome of a specified length.

    A genome is represented as a list of integers where each integer is randomly chosen
    from the set {0:Derecha, 1:Izquierda, 2:Arriba, 3:Abajo} and each integer represents a move.

    Parameters:
    genome_length (int): The length of the genome to be generated.

    Returns:
    list: A list of integers representing the genome.
    """
    return [random.randint(0, 3) for _ in range(genome_length)]


def generate_population(size, genome_length):
    """
    Generates a population of genomes.

    Args:
        size (int): The number of genomes in the population.
        genome_length (int): The length of each genome.

    Returns:
        list: A list of genomes, where each genome is a list of genes.
    """
    return [generate_genome(genome_length) for _ in range(size)]

def reward(individuo, maze, start, end):
    """
    Calculate the reward for a given individual navigating through a maze.
    Parameters:
    individuo (list of int): A list of moves where each move is represented by an integer (0: right, 1: left, 2: up, 3: down).
    maze (list of list of int): A 2D list representing the maze where 0 is an empty cell and 1 is a wall.
    start (tuple of int): The starting coordinates (x, y) in the maze.
    end (tuple of int): The goal coordinates (x, y) in the maze.
    Returns:
    tuple: A tuple containing the final score (int) and a set of visited cells (set of tuples).
    """
    visited = set()
    x, y = start
    score = 0
    last_move = None  # Mantiene el registro del último movimiento
    min_distance_to_goal = float('inf')  # Mantener la menor distancia lograda al objetivo

    for move in individuo:
        # Actualizar la posición según el movimiento
        prev_x, prev_y = x, y  # Guardar la posición anterior para comparación

        if move == 0:  # Derecha
            x += 1
        elif move == 1:  # Izquierda
            x -= 1
        elif move == 2:  # Arriba
            y -= 1
        elif move == 3:  # Abajo
            y += 1

        # Penalización por movimientos ilegales
        if x < 0 or y < 0 or x >= len(maze[0]) or y >= len(maze) or maze[y][x] == 1:
            score -= 500000  # Penalización ajustada
            break

        # Penalización por quedarse quieto
        if (y, x) == (prev_y, prev_x):
            score -= 1000000  # Penalización por no avanzar
            
        # Penalización por retroceder (hacer el movimiento opuesto al último)
        if last_move is not None:
            if (last_move == 0 and move == 1) or (last_move == 1 and move == 0) or \
               (last_move == 2 and move == 3) or (last_move == 3 and move == 2):
                score -= 80000  # Penalización por retroceder

        last_move = move  # Actualizar el último movimiento

        # Recompensa por visitar nuevas celdas
        if (y, x) not in visited:
            visited.add((y, x))
            score += 1000000
            if maze[y][x] == 0:
                score += 500000000  # Recompensa adicional por visitar celdas vacías

        # Penalización por visitar celdas ya visitadas
        if (y, x) in visited:
            score -= 500000

        
        #Penalización por alejarse o acercarse del objetivo
        distance_to_goal = abs(x - end[0]) + abs(y - end[1])
        score += max(0, 500000 - distance_to_goal)  # Ajustado para que sea menos abrupto
        
        # Gran recompensa si se alcanza el objetivo
        if (y, x) == end:
            score += 400000000000
            break

    return score, visited


def select(population, fitnesses, k):
    """
    Perform tournament selection on a population.
    Args:
        population (list): A list of individuals in the population.
        fitnesses (list): A list of fitness values corresponding to each individual in the population.
        k (int): The number of individuals to participate in each tournament.
    Returns:
        list: A list of selected individuals after performing tournament selection.
    """
   
    selected = []
    for _ in range(len(population)):
        tournament = random.sample(list(zip(population, fitnesses)), k)
        winner = max(tournament, key=lambda x: x[1])  # Selecciona el de mayor fitness
        selected.append(winner[0])
    return selected


def crossover(parent1, parent2, start, end):
    """
    Perform a crossover operation between two parent genomes to produce two child genomes.
    Args:
        parent1 (list): The first parent genome.
        parent2 (list): The second parent genome.
        start (int): The starting index for the crossover operation.
        end (int): The ending index for the crossover operation.
    Returns:
        tuple: A tuple containing two child genomes (child1, child2) resulting from the crossover.
    """
    
    genome_length = len(parent1)
    child1, child2 = [], []

    for i in range(genome_length):
        if random.random() < 0.5:  # 50% de probabilidad de heredar de parent1
            child1.append(parent1[i])
            child2.append(parent2[i])
        else:
            child1.append(parent2[i])
            child2.append(parent1[i])

    return child1, child2


def mutate(individual, mutation_rate, generation, max_generations):
    """
    Mutates an individual based on a dynamic mutation rate.
    The mutation rate is adjusted dynamically based on the current generation
    and the maximum number of generations. The dynamic mutation rate increases
    as the number of generations progresses.
    Parameters:
    individual (list): The individual to be mutated, represented as a list of integers.
    mutation_rate (float): The base mutation rate, a value between 0 and 1.
    generation (int): The current generation number.
    max_generations (int): The maximum number of generations.
    Returns:
    list: The mutated individual.
    """
   
    dynamic_rate = mutation_rate + (1 - (generation / max_generations)) * 0.2
    for i in range(len(individual)):
        if random.random() < dynamic_rate:
            individual[i] = random.randint(0, 3)
    return individual

def evolve(population, maze, start, end, generations=100, mutation_rate=0.1):
    """
    Evolves a population of individuals to solve a maze using a genetic algorithm.
    Args:
        population (list): The initial population of individuals.
        maze (list of list): The maze to be solved, represented as a 2D grid.
        start (tuple): The starting position in the maze (row, column).
        end (tuple): The ending position in the maze (row, column).
        generations (int, optional): The number of generations to evolve. Default is 100.
        mutation_rate (float, optional): The mutation rate for the genetic algorithm. Default is 0.1.
    Returns:
        list: The final evolved population.
    """
    best_visited_positions = set()
    best_fitness_overall = float('-inf')
    for generation in range(generations):
        # Se evaluar la aptitud de cada individuo
        fitnesses=[]
        visited_positions = []
        mejor_individuo = None
        for ind in population:
            score, visited = reward(ind, maze, start, end)
            fitnesses.append(score)
            visited_positions.append(visited)
        
        # Sacar mejor individuo
        best_fitness = max(fitnesses)
        best_individual_index = fitnesses.index(best_fitness)
        mejor_individuo = population[best_individual_index]

        # Selección
        selected_population = select(population, fitnesses, k=5)

        # Cruce
        next_generation = []
        for i in range(0, len(selected_population), 2):
            parent1 = selected_population[i]
            if i + 1 < len(selected_population):
                parent2 = selected_population[i + 1]
            else:
                parent2 = selected_population[0]
            child1, child2 = crossover(parent1, parent2, start, end)
            next_generation.extend([child1, child2])

        # Mutación
        next_generation = [mutate(ind, mutation_rate, generation, 1000) for ind in next_generation]
        
        # Actualizar población
        population = next_generation

        # Monitoreo
        best_fitness = max(fitnesses)
        best_individual_index = fitnesses.index(best_fitness)
        
        # Si encontramos un mejor fitness, actualizamos y graficamos
        if best_fitness > best_fitness_overall:
            best_fitness_overall = best_fitness
            best_visited_positions = visited_positions[best_individual_index]
            print(f"Generación {generation + 1}: Nuevo mejor fitness = {best_fitness}, mejor individuo = {mejor_individuo}, visitados = {best_visited_positions}")
            plot_solution_path_fixed(maze, start, end, best_visited_positions, generation + 1, best_fitness)
           
    return population
````
